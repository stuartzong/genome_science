```python
f = '/projects/trans_scratch/validations/workspace/szong/Cervical/integration/Cervical_transcriptome_integration_summary_20190625.txt'
df = pd.read_csv(f, sep='\t')
df.head(2)
```

```python
df['hpv'] = df['virus'].apply(lambda x: x.split('|')[4].split('_')[4].replace(',', ''))
```

```python
df = df[df.hpv.str.contains('16')]
```

```python
def unique_items(x):
    return ';'.join(x.unique())
```

```python
df1 = df.groupby('library')[['human-breakpoint', 'viral-breakpoint', 'hpv']].agg({'viral-breakpoint':['min', 'max'], 'hpv':[','.join, unique_items], 'human-breakpoint':','.join}).reset_index()
```
