```python
dfm = pd.melt(df, id_vars='Biospecimen ID', value_vars=['X coverage (incl. duplicate reads)',
       '% of reads on target (from total reads)',
       '% of reads off target (from total reads)', 'Raw Bases Sequenced (Gb)',
       'Number of bases Aligned (Gb)',
       'Number of bases with zero coverage (kb)'], value_name='value')
```

```python
cls = ['X coverage (incl. duplicate reads)',
       '% of reads on target (from total reads)',
       '% of reads off target (from total reads)', 'Raw Bases Sequenced (Gb)',
       'Number of bases Aligned (Gb)',
       'Number of bases with zero coverage (kb)']
```

```python
sns.boxplot(x='variable', y='value', data=dfm, fliersize=2.5)
```
